```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world-happiness-report-2021.csv')
df.head()
```

```python
# Renaming the Column

df.rename(columns = {'Ladder score':'Happiness score'}, inplace = True)
```

```python
df.head()
```

```python
# Making new dataframe of useful columns

raw_data = df[['Country name','Regional indicator','Happiness score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']].copy()
raw_data
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

### Relation between happiness score and GDP

```python
plt.figure(figsize=(16,8))
plt.title('Plot between Happiness score and GDP', fontsize=18, color='Blue')
sns.scatterplot(x='Happiness score', y='Logged GDP per capita', data=raw_data, hue='Regional indicator', s=200)
plt.show()
```

#### Inference : 

### Finding GDP per region

```python
gdp_region = raw_data.groupby('Regional indicator')['Logged GDP per capita'].sum()
gdp_region
```

```python
plt.figure(figsize =(12,8))
plt.title('GDP per Region', fontsize=18, color='Blue')
gdp_region.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

### See the Total counries

```python
total_countries = raw_data.groupby('Regional indicator')['Country name'].count()
print(total_countries)
```

```python
plt.figure(figsize =(12,8))
plt.title('Total Countries per Region', fontsize=18, color='Blue')
# plt.bar(x=total_countries.column, y=total_countries.index)
# sns.barplot(x=total_countries.column, y=total_countries.index, data=total_countries)
total_countries.plot.bar()
plt.xlabel('Regions', fontsize=14)
plt.ylabel('Total Countries', fontsize=14)
plt.xticks(rotation=70)
plt.show()
```

#### Inference : 

### Corruption in various regions

```python
Corrupt_regions = raw_data.groupby('Regional indicator')['Perceptions of corruption'].sum()
Corrupt_regions
```

```python
plt.figure(figsize =(12,8))
plt.title('Corruption in various Regions', fontsize=18, color='Blue')
Corrupt_regions.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

### Find relation between Freedom to make life choices and Happiness score	

```python
plt.figure(figsize=(16,8))
plt.title('Plot between Freedom to make life choices and Happiness score', fontsize=18, color='Blue')
sns.scatterplot(x='Happiness score', y='Freedom to make life choices', data=raw_data, hue='Regional indicator', s=200)
plt.xlabel('Happiness score', fontsize=14)
plt.ylabel('Freedom to make life choices', fontsize=14)
plt.show()
```

#### Inference :

### Countries with Least and Highest Corruption

```python
country_least_corr = raw_data.sort_values(by='Perceptions of corruption').head(10)
country_high_corr = raw_data.sort_values(by='Perceptions of corruption').tail(10)
```

```python
plt.figure(figsize =(16,5))
plt.subplot(1,2,1)
plt.title('Top 10 Countries with Least Corruption', fontsize=16, color='Blue')
sns.barplot(x = country_least_corr['Country name'], y = country_least_corr['Perceptions of corruption'], data = country_least_corr)
# plt.bar(x=country_least_corr['Country name'], y=country_least_corr['Perceptions of corruption'], height=10)
plt.ylabel('Perceptions of corruption')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.title('Top 10 Countries with Highest Corruption', fontsize=16, color='Blue')
sns.barplot(x = country_high_corr['Country name'], y = country_high_corr['Perceptions of corruption'], data = country_high_corr)
# plt.bar(x=country_high_corr['Country name'], y=country_high_corr['Perceptions of corruption'], height=10)
plt.ylabel('Perceptions of corruption')
plt.xticks(rotation=70)

plt.show()
```
